```python
jakarta = pd.DataFrame(dfJ)
bogor = pd.DataFrame(dfB)
bekasi = pd.DataFrame(dfX)
```

```python
# Slicing Data Jakarta
jakarta = jakarta.iloc[0:121] # selanjutnya data di seleksi menggunakan iloc untuk menyisihkan data yang tidak diperlukan.
jakarta = jakarta[["created_at", "LT", "LB", "KT", "KM", "garasi", "carport", "lokasi", "sertifikat", "listrik", "hadap", "harga", "URL", "deskripsi"]]
jakarta = jakarta.fillna(0) # Untuk mengganti NaN dengan 0 maka digunakan fungsi fillna karena Nan tidak dapat diproses.
jakarta.info()

# Berdasarkan hasil output terdapat 121 entries dengan 14 columns, serta terdapat 8 columns tipe float dan 6 columns tipe object/string.
# Karena variabel terikat yang dibutuhkan pada kasus ini sudah memiliki tipe data float, maka tidak perlu melakukan perubahan tipe data variabel.
```

```python
# Slicing Data Bogor
bogor = bogor.iloc[0:193] # selanjutnya data di seleksi menggunakan iloc untuk menyisihkan data yang tidak diperlukan.
bogor = bogor[["created_at", "LT", "LB", "KT", "KM", "garasi", "carport", "lokasi", "sertifikat", "listrik", "hadap", "harga", "URL", "deskripsi"]]
bogor = bogor.fillna(0) # Untuk mengganti NaN dengan 0 maka digunakan fungsi fillna karena Nan tidak dapat diproses.
bogor.info()

# Berdasarkan hasil output terdapat 193 entries dengan 14 columns, serta terdapat 8 columns tipe float dan 6 columns tipe object/string.
# Karena variabel terikat yang dibutuhkan pada kasus ini sudah memiliki tipe data float, maka tidak perlu melakukan perubahan tipe data variabel.
```

```python
# Slicing Data Bekasi
bekasi = bekasi.iloc[0:107] # selanjutnya data di seleksi menggunakan iloc untuk menyisihkan data yang tidak diperlukan.
bekasi = bekasi[["created_at", "LT", "LB", "KT", "KM", "garasi", "carport", "lokasi", "sertifikat", "listrik", "hadap", "harga", "URL", "deskripsi"]]
bekasi = bekasi.fillna(0) # Untuk mengganti NaN dengan 0 maka digunakan fungsi fillna karena Nan tidak dapat diproses.
bekasi.info()

# Berdasarkan hasil output terdapat 107 entries dengan 14 columns, serta terdapat 8 columns tipe float dan 6 columns tipe object/string.
# Karena variabel terikat yang dibutuhkan pada kasus ini sudah memiliki tipe data float, maka tidak perlu melakukan perubahan tipe data variabel.
```

```python
# Mencari Duplicate Data
print(jakarta.duplicated().sum())
print(bogor.duplicated().sum())
print(bekasi.duplicated().sum())
# Berdasarkan hasil output tidak ditemukan duplicate data pada semua database
```

```python
df = jakarta.merge(bogor, how='outer')
df = df.merge(bekasi, how='outer')
df.info()
```

```python
# Statistik Deskriptif Data
print(f'Jumlah baris sebelum memfilter outlier: {len(df)}')
#kolom = [["LT", "LB", "KT", "KM", "garasi", "carport", "listrik", "harga"]]
for col in ['LT', 'LB', 'KT', 'KM', 'garasi', 'carport', 'listrik', 'harga']:
  Q1 = df[col].quantile(0.25) 
  Q3 = df[col].quantile(0.75)
  IQR = Q3-Q1
  low_limit = Q1 - 1.5*IQR
  high_limit = Q3 + 1.5*IQR
  df = df[(df[col] >= low_limit) & (df[col] <= high_limit)]
print(f'Jumlah baris setelah memfilter outlier: {len(df)}') # Untuk menghitung banyaknya baris yang sudah di outlier digunakan fungsi len.
# Berdasarkan hasil output didapatkan 182 baris dari 421 baris yang sudah di outlier.
```

```python
print(f'Jumlah baris sebelum memfilter outlier: {len(jakarta)}')
#kolom = [["LT", "LB", "KT", "KM", "garasi", "carport", "listrik", "harga"]]
for col in ['LT', 'LB', 'KT', 'KM', 'garasi', 'carport', 'listrik', 'harga']:
  Q1 = jakarta[col].quantile(0.25) 
  Q3 = jakarta[col].quantile(0.75)
  IQR = Q3-Q1
  low_limit = Q1 - 1.5*IQR
  high_limit = Q3 + 1.5*IQR
  jakarta = jakarta[(jakarta[col] >= low_limit) & (jakarta[col] <= high_limit)]
print(f'Jumlah baris setelah memfilter outlier: {len(jakarta)}')
```

```python
print(f'Jumlah baris sebelum memfilter outlier: {len(bekasi)}')
#kolom = [["LT", "LB", "KT", "KM", "garasi", "carport", "listrik", "harga"]]
for col in ['LT', 'LB', 'KT', 'KM', 'garasi', 'carport', 'listrik', 'harga']:
  Q1 = bekasi[col].quantile(0.25) 
  Q3 = bekasi[col].quantile(0.75)
  IQR = Q3-Q1
  low_limit = Q1 - 1.5*IQR
  high_limit = Q3 + 1.5*IQR
  bekasi = bekasi[(bekasi[col] >= low_limit) & (bekasi[col] <= high_limit)]
print(f'Jumlah baris setelah memfilter outlier: {len(bekasi)}')
```
